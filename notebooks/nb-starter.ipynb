```python
import altair as alt
import altair_viewer
import numpy as np
import pandas as pd
from pyspark.sql import DataFrame
from vega import VegaLite
```

```python
df: DataFrame = spark.createDataFrame(np.random.randn(200, 3).tolist(), ["a", "b", "c"])
pdf: pd.DataFrame = df.toPandas()
```

```python
spec = {
    "mark": {"type": "circle", "tooltip": True},
    "encoding": {
        "x": {"field": "a", "type": "quantitative"},
        "y": {"field": "b", "type": "quantitative"},
        "size": {"field": "c", "type": "quantitative"},
        "color": {"field": "c", "type": "quantitative"},
    },
}
altair_viewer.display(VegaLite(spec, pdf).spec, inline=True)
```

```python
spec = {
    "mark": {"type": "bar", "tooltip": True},
    "encoding": {
        "x": {"field": "a", "type": "nominal"},
        "y": {"field": "b", "type": "quantitative"},
    },
    "data": {
        "values": [
            {"a": "A", "b": 28},
            {"a": "B", "b": 55},
            {"a": "C", "b": 43},
            {"a": "D", "b": 91},
            {"a": "E", "b": 81},
            {"a": "F", "b": 53},
            {"a": "G", "b": 19},
            {"a": "H", "b": 87},
            {"a": "I", "b": 52},
        ]
    },
}

altair_viewer.display(spec, inline=True)
```

```python
source = pd.DataFrame(
    {
        "a": ["A", "B", "C", "D", "E", "F", "G", "H", "I"],
        "b": [28, 55, 43, 91, 81, 53, 19, 87, 52],
    }
)

chart = alt.Chart(source).mark_bar().encode(x="a", y="b")
altair_viewer.display(chart, inline=True)
```
